```python
dados.rename(columns={"checking_status":'Checando Status', "duration": 'Duração', 'credit_history': "Histórico", 'purpose':'Propósito','credit_amount': 'Valor do Crédito','savings_status':'Situação da Poupança','employment':'Tempo de Emprego',"installment_commitment" :"Valor da Parcela Cometida", 'personal_status':"Status Social","other_parties": 'Outros Envolvidos', "property_magnitude" :'Propriedade', "age":"Idade","other_payment_plans": "Outros Planos de Pagamento","housing":"Moradia", "existing_credits": "Creditos Existentes", "job": "Trabalho", "num_dependents":"Dependentes","own_telephone":"Telefone Próprio", "foreign_worker": "Estangeiro", "class": "Score" }, inplace = True)
```

```python
dados.head()
```

```python
dados.isnull().sum()
```

```python
for col in dados.columns:
    if dados[col].dtype == "object":
        print(f"\n{col}:\n{dados[col].unique()}")
```

```python
for col in dados.columns:
    if dados[col].dtype == "object":
        dados[col] = dados[col].str.replace("'", "")
```

```python
dados['Checando Status'] = dados['Checando Status'].map({
    'no checking': 'sem conta',
    '<0': 'menos de 0',
    '0<=X<200': 'entre 0 e 200',
    '>=200': 'acima de 200'
})
```

```python
dados.head()
```

```python
dados['Histórico'] = dados['Histórico'].map({
    'no credits/all paid': 'sem crédito/tudo pago',
    'all paid': 'tudo pago',
    'existing paid': 'crédito existente pago',
    'delayed previously': 'atraso anterior',
    'critical/other existing credit': 'crédito crítico ou outro'
})
```

```python
dados['Propósito'] = dados['Propósito'].map({
    'radio/tv': 'rádio/tv',
    'education': 'educação',
    'furniture/equipment': 'móveis/equipamento',
    'new car': 'carro novo',
    'used car': 'carro usado',
    'business': 'negócios',
    'domestic appliance': 'eletrodoméstico',
    'repairs': 'reparos',
    'other': 'outros',
    'retraining': 'reciclagem profissional'
})
```

```python
dados['Situação da Poupança'] = dados['Situação da Poupança'].map({
    'no known savings': 'sem poupança conhecida',
    '<100': 'menos de 100',
    '500<=X<1000': 'entre 500 e 1000',
    '>=1000': 'mais de 1000',
    '100<=X<500': 'entre 100 e 500'
})
```

```python
dados['Tempo de Emprego'] = dados['Tempo de Emprego'].map({
    '<1': 'menos de 1 ano',
    '1<=X<4': 'entre 1 e 4 anos',
    '4<=X<7': 'entre 4 e 7 anos',
    '>=7': '7 anos ou mais',
    'unemployed': 'desempregado'
})
```

```python
dados['Status Social'] = dados['Status Social'].map({
    'male single': 'homem solteiro',
    'female div/dep/mar': 'mulher (divorciada, dependente ou casada)',
    'male div/sep': 'homem divorciado/separado',
    'male mar/wid': 'homem casado/viúvo'
})
```

```python
dados['Outros Envolvidos'] = dados['Outros Envolvidos'].map({
    'none': 'nenhum',
    'guarantor': 'fiador',
    'co applicant': 'co-requerente'
})
```

```python
dados['Propriedade'] = dados['Propriedade'].map({
    'car': 'carro',
    'life insurance': 'seguro de vida',
    'no known property': 'sem propriedade conhecida',
    'real estate': 'imóvel'
})
```

```python
dados['Outros Planos de Pagamento'] = dados['Outros Planos de Pagamento'].map({
    'none': 'nenhum',
    'bank': 'banco',
    'stores': 'lojas'
})
```

```python
dados['Moradia'] = dados['Moradia'].map({
    'own': 'própria',
    'for free': 'gratuita',
    'rent': 'alugada'
})
```

```python
dados['Trabalho'] = dados['Trabalho'].map({
    'skilled': 'qualificado',
    'unskilled resident': 'não qualificado (residente)',
    'high qualif/self emp/mgmt': 'altamente qualificado/autônomo/gestão',
    'unemp/unskilled non res': 'desempregado/não qualificado (não residente)'
})
```

```python
dados['Telefone Próprio'] = dados['Telefone Próprio'].map({
    'yes': 'sim',
    'no': 'não'
})
```

```python
dados['Estangeiro'] = dados['Estangeiro'].map({
    'yes': 'sim',
    'no': 'não'
})
```

```python
dados['Score'] = dados['Score'].map({
    'good': 'bom',
    'bad': 'ruim'
})
```

```python
# Conferência geral das colunas categóricas
for col in dados.columns:
    if dados[col].dtype == "object":
        print(f"\n{col}:\n{dados[col].unique()}")
```
